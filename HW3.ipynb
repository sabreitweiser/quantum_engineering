```python
%matplotlib inline 
import matplotlib.pyplot as plt
import numpy as np
from qutip import *
```

```python
N=2
up = basis(N,0)
dn = basis(N,1)
psiz = up 
psix = (up+dn)/np.sqrt(2) #+,X eigenstate
psiy = (up+1j*dn).unit() #+,Y eigenstate
```

```python
def show_bloch(result_to_plot, psi):
    sphere = Bloch()
    sphere.add_vectors([0, 0, 1]) # Add a vector showing the initial orientation
    sphere.add_points(result_to_plot.expect,meth='l') # Draw the calculated spin projections as a line trajectory
    sphere.show()
```

```python
def calc_evol(psi, c_ops):
    tlist = np.linspace(0.0,10*np.pi,1000)
    result = mesolve(sigmaz(),psi,tlist,c_ops,[sigmax(),sigmay(),sigmaz()])
    show_bloch(result, psi)
```

```python
calc_evol((psix + psiz).unit(), [10*sigmax()])
```
